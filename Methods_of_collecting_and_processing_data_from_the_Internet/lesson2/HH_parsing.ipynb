```python
import bs4
import requests
import lxml
import pandas as pd
import re
```

```python
def parsing(vacancy):
    params = {
        'text': vacancy,
        'search_field': 'name',
        'area': '',
        'salary': '',
        'currency_code': 'RUR',
        'experience': 'doesNotMatter',
        'order_by': 'relevance',
        'search_period': '0',
        'items_on_page': '100',
        'no_magic': 'true',
        'L_save_area': 'true'
    }

    headers = {
        'User-Agent': 'Mozilla/5.0 (X11; Linux x86_64; rv:69.0) Gecko/20100101 Firefox/69.0'
    }
    url = 'https://hh.ru/search/vacancy/'
    
    data = requests.get(url, params=params, headers=headers)
    print(str(data))
    if str(data) == '<Response [200]>':
        soup = bs4.BeautifulSoup(data.text, 'html.parser')
        return soup
    else:
        return None
```

```python
def processing(soup):
    vacancies = soup.find_all('div', {"class":"vacancy-serp-item"})
    vacancies_list = []
    for vacancy in vacancies:
        vacancy_dict = {}
        
        # name
        name = vacancy.find("span", {"class" : "g-user-content"}).text
        vacancy_dict['name'] = name
        
        # salary
        salary = vacancy.find('div', {'class': 'vacancy-serp-item__sidebar'})
        if not salary:
            salary = None
        else:
            vacancy_dict['salary'] = salary.text

        # link
        link = vacancy.find("a", {"data-qa" : "vacancy-serp__vacancy-title"})['href']
        vacancy_dict['link'] = link
        
        # source
        vacancy_dict['source'] = 'hh.ru'
        vacancies_list.append(vacancy_dict)
    return vacancies_list
```

```python
soup = parsing('фронтенд')
```

```python
vacancies_list = processing(soup)
df = pd.DataFrame(vacancies_list, columns=['name', 'salary', 'link', 'source'])
df
```

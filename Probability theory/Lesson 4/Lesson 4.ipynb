Случайная непрерывная величина A имеет равномерное распределение на промежутке (200, 800].<br> Найдите ее среднее значение и дисперсию.

Непрерывное равномерное распределение (`scipy.stats.uniform`) — непрерывный аналог дискретного равномерного распределения. Функция плотности постоянна внутри отрезка:
$$
    f(x) = \begin{cases}
    1 / (b - a), & x \in [ a, b ], \\
    0 & \text{иначе}.
    \end{cases}
$$

```python
m, d = uniform(loc=200, scale=600).stats()
print( f'Среднее значение: {m}\nДисперсия: {d}')
```

Задача 3  
Непрерывная случайная величина X распределена нормально и задана плотностью распределения  
$$
   \Large f(x) = \frac{1}{4 \sqrt {2\pi}}e^\frac{-\left({x+2}\right)^{2}}{32}
$$

Найдите:

а). M(X)

б). D(X)

в). std(X) (среднее квадратичное отклонение)

Функция плотности вероятности, совпадающая с функцией Гаусса:
$$
  \Large f(x)={\frac{1}{\sigma {\sqrt {2\pi }}}}e^{-{\frac {1}{2}}\left({\frac {x-\mu }{\sigma }}\right)^{2}}, 
$$ где параметр $\mu$  — математическое ожидание (среднее значение), медиана и мода распределения, а параметр $\sigma$ — среднеквадратическое отклонение, $\sigma ^{2}$ — дисперсия распределения.
Заданная функция с учетом функции плотности вероятности примет вид:
$$
   \Large f(x) = \frac{1}{4 \sqrt {2\pi}}e^{-{\frac {1}{2}}\left(\frac{\left({x - (-2)}\right)}{4}\right)^{2}}
$$
Следовательно, математическое ожидание $M(X) = -2$, среднеквадратическое отклонение $std(X) = 4$, а дисперсия $D(X) = 16$


Задача 4  
Рост взрослого населения города X имеет нормальное распределение.  
Причем, средний рост равен 174 см, а среднее квадратичное отклонение равно 8 см.  
Какова вероятность того, что случайным образом выбранный взрослый человек имеет рост:  
a). больше 182 см  
b). больше 190 см  
c). от 166 см до 190 см  
d). от 166 см до 182 см  
e). от 158 см до 190 см  
f). не выше 150 см или не ниже 190 см  
g). не выше 150 см или не ниже 198 см  
h). ниже 166 см.  

```python
letters = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h']
lst = [norm.sf(182, loc = 174, scale = 8),
            norm.sf(190, loc = 174, scale = 8),
            norm.cdf(190, loc = 174, scale = 8) - norm.cdf(166, loc = 174, scale = 8),
            norm.cdf(182, loc = 174, scale = 8) - norm.cdf(166, loc = 174, scale = 8),
            norm.cdf(190, loc = 174, scale = 8) - norm.cdf(158, loc = 174, scale = 8),
            norm.cdf(150, loc = 174, scale = 8) + norm.sf(190, loc = 174, scale = 8),
            norm.cdf(150, loc = 174, scale = 8) + norm.sf(198, loc = 174, scale = 8),
            norm.cdf(166, loc = 174, scale = 8)]
for i in range(len(lst)):
    print(f'{letters[i]}: {lst[i]}')
```

Задача 5  
На сколько сигм (средних квадратичных отклонений) отклоняется рост человека, равный 190 см, от
математического ожидания роста в популяции, в которой M(X) = 178 см и D(X) = 25 кв.см?


Для вычисления разброса значений нормально распределённой случайной величины можно использовать следующие правила:

* Интервал от $\mu - \sigma$ до $\mu + \sigma$ (__стандартное отклонение__) содержит около $68\%$ вероятностной массы (т.е. с вероятностью $68\%$ данная величина попадает в этот интервал).
* От $\mu - 2\sigma$ до $\mu + 2\sigma$ — около $95\%$ массы (__правило двух сигм__).
* От $\mu - 3\sigma$ до $\mu + 3\sigma$ — около $99.7\%$ массы (__правило трёх сигм__).

Для данной задачи $\mu = 178, \sigma = 5$. Значение x = 190 попадает в интервал заданный __правилом трех сигм__, т.е. $[163, 193]$, а точнее $2,4 \sigma$

```python
# Посчитайте производную функции y = x**3 + z - 75t в точке (1, 0.5, 2)

x = torch.tensor([1.0], requires_grad=True)
z = torch.tensor([0.5], requires_grad=True)
t = torch.tensor([2.0], requires_grad=True)
y = (x**3 + z - 75*t).sum()
print(y)
y.backward()
print(x.grad, z.grad, t.grad)
```
